```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./data/gas_02.csv')
```

```python
list_columns = ['Year', 'All', 'Seoul', 'Busan', 'Daegu', 'Incheon', 'Gwangju', 'Daejeon', 'Ulsan', 'Sejong', 'Gyeonggi', 'Gangwon', 'Chungbuk', 'Chungnam', 'Jeonbuk', 'Jeonnam', 'Gyeongbuk', 'Gyeongnam', 'Jeju']
```

```python
df.columns = list_columns
```

```python
for i in range(2, len(df.columns)) :
    num_all = df['All'][i]
    j = 0
    for j in range(0, len(df)) :
        if df[df[list_columns[i]].name][j] == 0 :
            continue
        else :
            df[df[list_columns[i]].name][j] = round(df[df[list_columns[i]].name][j] / num_all, 4)
```

```python
df.to_csv('./data/data_gas.csv', index=False)
```
