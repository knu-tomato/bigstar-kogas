```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## [1] 민수용, 산업용 컬럼 정리하기
---

```python
data = pd.read_csv("data_final.csv")

data.info()
```

```python
# 민수용, 산업용 컬럼 정리
civil = [
    "AT",
    "LT",
    "HT",
    "CAP_TEMP_EF",
    "DONGNAM_TEMP_EF",
    "DAEGYUNG_TEMP_EF",
    "CHOONG_TEMP_EF",
    "HONAM_TEMP_EF",
    "MEAN_TEMP_EF",
]
ind = [
    "POWER_WATER",
    "POWER_FIRE",
    "POWER_NUCLEAR",
    "POWER_NAEYEON",
    "POWER_GROUP",
    "POWER_TRANS",
    "POWER_GAS",
    "POWER_JUNG",
    "POWER_MOO",
    "POWER_YOU",
    "POWER_LNG",
    "all",
    "family",
    "public",
    "service",
    "civil_sum",
    "argri_fishing",
    "mining",
    "manufact",
    "ind_sum",
]
```

```python
data_civil = data.loc[:, civil]
data_ind = data.loc[:, ind]
```

```python
# 민수용 컬럼명 정리
civil_columns = [
    "TCL_AT",
    "TCL_LT",
    "TCL_HT",
    "TEMP_EF_CAP",
    "TEMP_EF_DONGNAM",
    "TEMP_EF_DAEGYUNG",
    "TEMP_EF_CHOONG",
    "TEMP_EF_HONAM",
    "TEMP_EF_MEAN",
]

# 산업용 컬럼명 정리
ind_columns = [
    "POWER_WATER",
    "POWER_FIRE",
    "POWER_NUCLEAR",
    "POWER_NAEYEON",
    "POWER_GROUP",
    "POWER_TRANS",
    "POWER_GAS",
    "POWER_JUNG",
    "POWER_MOO",
    "POWER_YOU",
    "POWER_LNG",
    "ELEC_ALL",
    "ELEC_FAMILY",
    "ELEC_PUBLIC",
    "ELEC_SERVICE",
    "ELEC_CIVIL_SUM",
    "ELEC_ARGRI_FISHING",
    "ELEC_MINING",
    "ELEC_MANUFACT",
    "ELEC_INDSUM",
]

data_civil.columns = civil_columns
data_ind.columns = ind_columns

# 년도 컬럼 추출 및 컬럼 이름 변경
data_date = pd.DataFrame(data.iloc[:, 0])
data_date.columns = ["DATE"]
```

## [2] 데이터 파일 생성
---

```python
# 종합 데이터 만들기
data_full = pd.concat([data_date, data_civil, data_ind], axis=1)

data_full.head()
```

```python
# 데이터 저장
data_full.to_csv("data_tomato.csv", index=False)
data_civil.to_csv("data_tomato_civil.csv", index=False)
data_ind.to_csv("data_tomato_ind.csv", index=False)
```
