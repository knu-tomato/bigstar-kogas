```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./data/gas.csv')
```

```python
df.columns
```

```python
df_01 = df.drop(['시점','호수 (호)'], axis=1)
```

```python
plt.figure(figsize = (12,14))
plt.plot(df_01)
plt.rc('font', family='D2Coding') 
plt.legend(df_01.columns)
plt.show()
```
