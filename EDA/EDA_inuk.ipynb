```python
data.columns
```

```python
temp_cols = [col for col in data.columns if "TEMP" in col]
tcl_cols = [col for col in data.columns if "TCL" in col]
power_cols = [col for col in data.columns if "POWER" in col]
elec_cols = [col for col in data.columns if "ELEC" in col]

data_temp = data[temp_cols]
data_tcl = data[tcl_cols]
data_power = data[power_cols]
data_elec = data[elec_cols]

data_qva = data["QVA"]
data_rv = data["RV"]
data_civil = data["CIVIL"]
data_ind = data["IND"]
```

```python
data_power.shape
```

```python
def draw_plots(data, figsize, color):
    len = data.shape[1]
    data_cols = data.columns
    date = pd.to_datetime(data.index)

    cols = 4
    rows = (len // 5) + 1

    plt.figure(figsize=figsize)

    for i, data_col in zip(range(1, len), data_cols):
        plt.subplot(rows, cols, i)
        plt.plot(date, data[data_col], color=color)
        plt.title(data_col)
    
    plt.show()
```

```python
draw_plots(data_temp, (20, 10), "skyblue")
```

```python
draw_plots(data_tcl, (20, 5), "tomato")
```

```python
draw_plots(data_elec, (20, 10), "orange")
```

```python
draw_plots(data_power, (20, 15), 'green')
```
